```python
plt.xkcd()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([])
plt.yticks([])
ax.set_ylim([-50, 20])

data = np.ones(100)
data[60:] -= np.arange(40)

plt.annotate(
    'trying to decide whether to code or make music',
    xy=(60, 1), arrowprops=dict(arrowstyle='->'), xytext=(40, 10))

plt.plot(data, 'g-')

plt.xlabel('time')
plt.ylabel('hours in day')
```

```python
#remove xkcd mode:
plt.rcdefaults()

values = [12, 55, 4, 25, 22]
colors = ['g', 'r', 'b', 'c', 'm']
explode = [0, 0, 0.4, 0, 0]
labels = ['Mexico', 'United States', 'Canada', 'China', 'Europe']
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Large and Incharge')
plt.show()
```

```python
randn = np.random.randn
X = randn(500)
Y = randn(500)
plt.scatter(X,Y)
plt.show()
```

```python
incomes = np.random.normal(38888, 1999, 10000)
plt.hist(incomes, 50)
plt.show()
```

```python
uniformSkewed = randn(100) * 100 -40
high_outliers = randn(10) * 50 + 100
low_outliers = randn(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```
